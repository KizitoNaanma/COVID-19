```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('covid_19_data.csv')
```

```python
data.head()
```

```python
data = data.drop(['Province/State', 'SNo', 'Last Update'], axis=1)
```

```python
data = data.rename(columns={'Country/Region': 'Country', 'ObservationDate':'Date'})
```

```python
data.head()
```

```python
data.groupby('Date').describe()
```

```python
data.groupby('ObservationDate')['Confirmed'].sum()
```

```python
mod_data = data[data.Date == data.Date.max()]
mod_data = mod_data.groupby('Country').sum().sort_values(by='Confirmed', ascending = False)
mod_data.head(10)
```

```python
mod_data['Death Rate'] = mod_data['Deaths'] / mod_data['Confirmed'] * 100
mod_data['Recovery Rate'] = mod_data['Recovered'] / mod_data['Confirmed'] * 100
mod_data['Active'] = mod_data['Confirmed'] - mod_data['Deaths'] - mod_data['Recovered']
```

```python
mod_data = mod_data.reset_index()
mod_data.head()
```

```python
mod_data.reset_index()
```

```python
mod_head = mod_data.head(20)
mod_tail = mod_data.tail(20)
```

```python
fig = px.pie(mod_data, values='Confirmed', names= 'Country',
             title='Infected rates across countries',
              hover_data=['Country'])
fig.update_traces(textposition='inside')
fig.show()
```

```python
# fig = px.bar(mod_head, x=case, y='Confirmed')
# fig.show()
fig = px.bar(mod_head, x='Country', y='Confirmed',
             hover_data=['Death Rate', 'Recovery Rate'], color='Active')
fig.show()
```

```python
fig = px.bar(mod_tail, x='Country', y='Confirmed',
             hover_data=['Death Rate', 'Recovery Rate'], color='Active')
fig.show()
```
